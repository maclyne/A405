```python
%matplotlib inline

from matplotlib import pyplot as plt
import numpy as np
```

```python
# This notebook draws figure 1.9 from Thompkins (a Tephigram)

P_o = 1000 * 10**2 # reference height (Pa)
Rd = 287.05 # J Kg^-1 K^-1
c_p = 1005 # J Kg^-1 K^-1 (dry air)
title = 'Tephigram'
plt.style.use('ggplot')
fig,ax = plt.subplots(1,1,figsize=(10,8))
xvals=np.linspace(270.,380.)

ax.plot(xvals,np.log(xvals * ((P_o / 100000 )**(Rd / c_p))),label='1000 hPa')
ax.plot(xvals,np.log(xvals * ((P_o / 80000 )**(Rd / c_p))),label='800 hPa') 
ax.plot(xvals,np.log(xvals * ((P_o / 60000 )**(Rd / c_p))),label='600 hPa') 


ax.legend()
out=ax.set(xlabel="T (K)",ylabel="Ln(potential temp)",title =title)
```
